```python
import requests
import time

def fetch_spot_order_book(symbol):
    """
    Obtém os dados de profundidade do order book para Spot.
    """
    url = f"https://api.mexc.com/api/v3/depth?symbol={symbol.replace('_', '')}&limit=5"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        bids = data["bids"]
        asks = data["asks"]

        if bids and asks:
            best_bid = float(bids[0][0])  # Melhor preço de compra
            best_ask = float(asks[0][0])  # Melhor preço de venda
            return best_bid, best_ask
        else:
            return None, None
    except Exception as e:
        print(f"[ERROR] Falha ao obter dados Spot para {symbol}: {e}")
        return None, None

def fetch_futures_order_book(symbol):
    """
    Obtém os dados de profundidade do order book para Futures.
    """
    url = f"https://contract.mexc.com/api/v1/contract/depth/{symbol}"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()['data']
        bids = data["bids"]
        asks = data["asks"]

        if bids and asks:
            best_bid = float(bids[0][0])  # Melhor preço de compra
            best_ask = float(asks[0][0])  # Melhor preço de venda
            return best_bid, best_ask
        else:
            return None, None
    except Exception as e:
        print(f"[ERROR] Falha ao obter dados Futures para {symbol}: {e}")
        return None, None

def calculate_spread_percentage(price1, price2):
    """
    Calcula o spread percentual entre dois preços.
    """
    average_price = (price1 + price2) / 2
    spread_percentage = abs(price1 - price2) / average_price * 100
    return spread_percentage

def check_arbitrage_opportunity(symbol, threshold=1):
    """
    Verifica oportunidades de arbitragem e emite alertas.
    """
    # Obter dados do Spot
    spot_bid, spot_ask = fetch_spot_order_book(symbol)
    # Obter dados do Futures
    futures_bid, futures_ask = fetch_futures_order_book(symbol)

    if spot_bid is None or spot_ask is None or futures_bid is None or futures_ask is None:
        print("[WARNING] Dados insuficientes para calcular o spread.")
        return

    # Calcular spreads
    spread_spot_to_futures = calculate_spread_percentage(spot_ask, futures_bid)
    spread_futures_to_spot = calculate_spread_percentage(futures_ask, spot_bid)

    # Emitir alertas
    if spread_spot_to_futures > threshold:
        print(f"[ALERTA] Spread Spot para Futures = {spread_spot_to_futures:.4f}%")
        print(f"Venda em Spot a {spot_ask}, Compre em Futures a {futures_bid}.")
    elif spread_futures_to_spot > threshold:
        print(f"[ALERTA] Spread Futures para Spot = {spread_futures_to_spot:.4f}%")
        print(f"Venda em Futures a {futures_ask}, Compre em Spot a {spot_bid}.")
    else:
        print(f"[INFO] Nada . Spot-Futures: {spread_spot_to_futures:.2f}%, Futures-Spot: {spread_futures_to_spot:.2f}%. - spot = {spot_bid, spot_ask} | futuro = {futures_ask, spot_bid} | {symbol} ")

def main():
    """
    Executa verificações contínuas de arbitragem para uma lista de símbolos.
    """
    symbols = ["APT_USDT", "NEAR_USDT", "TIA_USDT","POPCAT_USDT","WLD_USDT","PNUT_USDT","INJ_USDT", "TAO_USDT",
              "GOAT_USDT", "FET_USDT", "SEI_USDT", "RUNE_USDT", "NOT_USDT", "KAS_USDT","DOGS_USDT", "BOME_USDT", "CHILLGUY_USDT",
              "ORDI_USDT","TRX_USDT" ,"JUP_USDT", "NEIROCTO_USDT", "ARKM_USDT", "THE_USDT", "NEIROETH_USDT", "MYRO_USDT", "PONKE_USDT",
              "MEW_USDT"]  # Substitua pelos ativos desejados
    while True:
        for symbol in symbols:
            check_arbitrage_opportunity(symbol, threshold=1)
        time.sleep(1)  # Intervalo de 1 segundo entre verificações    # vende futuro 0.3056 e compra spot a 0.3024

if __name__ == "__main__":
    main()
```
